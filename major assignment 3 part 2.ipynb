```python
import pandas as pd 
import networkx as nx
import json
```

```python
df=pd.read_csv("https://github.com/umassdgithub/Week-8-part-1/blob/main/data/data_scopus.csv?raw=true")
```

```python
nodes=df['EID'].values
```

```python
G=nx.Graph()
```

```python
df.head()
```

```python
G.add_nodes_from([])
```

```python
nodes=[]
for row in df.iterrows():
    nodes.append((row[1]['EID'],{
        "Title":row[1]['Title']
    }))
```

```python
G.add_nodes_from(nodes)
```

```python
edges=[]
for row in df.iterrows():
#     current_paper=row[1]['EID']

    current_paper=row[1]['Author(s) ID']
    if ";" in row[1]['Author(s) ID']:
        authors=row[1]['Author(s) ID'][:-1].split(";")
        for author in authors:
            papers=df[df['Author(s) ID'].str.contains(author)]['EID'].values
        
            if len(papers)>0:
                for paper in papers:
                    if paper!= current_paper:
                        edges.append((current_paper,paper))
    
    
    
    else:
        continue
        
```

```python
G.add_edges_from(edges)
```

```python
pos=nx.drawing.spiral_layout(G)
```

```python
nx.draw(G,pos=pos,node_size=40)
```

```python
from networkx.readwrite import json_graph
```

```python
with open('publication_network.json','w') as f:
    json.dump(json_graph.node_link_data(G),f)
```
